```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r'c:\Users\sanju\Desktop\sample_Data.csv')
df.head(5)
```

```python
plt.scatter(df['Horsepower'], df['Engine_size'])
```

```python
wcss= []

for i in range(1,11):
    km= KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)

wcss
```

```python
plt.plot(range(1,11), wcss)
```
