Machine learning only deals with interger values. So to deal with catergorical and non-numeric data we convert them into specific numeric value using interger encoding or label encoding. But in such cases after converting, there will not be prbolem with ordinal variables (varibles have certain order) like ML model might think those integer values in order. like in case of degreee of education (bachelors then masters then PhD). In case of nominal variables, there will e problem as these variables doesn't have certain ordering. In such conditions, we use one hot encoding method.    

For ordinal variable we use ordinal encoding. And for target column that is y column, we use label encoding

Here we create a new column for each of out categories (nominal varaibles) present in the orginal dataset. then we asign binary 0 and 1 to each value

**1. One Hot Encoding Using Pands**

```python
dummy_df = pd.get_dummies(df['town'], dtype= int)
dummy_df
```

```python
merged_df = pd.concat([df, dummy_df], axis=1)
merged_df
```

Drop the nominal variable column as we don't need it anymoe along with one dummy varible column to avoid 'dummy varible trap'.
It happens if a varible can be determined from other variables. This trap causes multicolinearity which doesn't work with linear regression model. If we are using sklearn for LR, then no need to drop dummy variable column. 

**Multicollinearity:**  if independepent variabes have relationship among them then they becomes dependent. Input variables supposed to be independent. This is multicollinearity. As this multicolinearity comes after creating dummy varibles, it is called dummy variable traps. To avoid this, we remove one dummy variable column out of n numbers of column.  

```python
final_df = merged_df.drop(['town', 'west windsor'], axis= 1)
```

**2. Using sklearn library**

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(dtype='int')# drop= 'first' # putting drop= First argument to drop first dummy column to avoid dummy variable trap.
# put argument sparse = False in OneHotEncoder() to avoid creating sparse matrix and using toarray().
```

```python
# now we need to encode the nominal column and we have to join it to the x train column again.
df_temp = ohe.fit_transform(df[['town']]).toarray() # storing it to array otherwise it stores as a sparse matrix
df_temp # 3 columns as 3 different categories are there.
```

```python
# changing it to a dataframe
df_temp_1 = pd.DataFrame(df_temp, columns= ['monroe township', 'west windsor', 'robinsville'])
df_temp_1
```

```python
# joining 
df_concat = pd.concat([df, df_temp_1], axis= 1)
df_concat
```

```python
df_final = df_concat.drop(df_concat[['town', 'robinsville']], axis= 1)
```
